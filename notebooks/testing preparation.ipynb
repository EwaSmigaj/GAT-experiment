# Testing Preparation: Split data to be able to run tests

Currently, I don't have the compute power to run tests, so in odrer to perform testing, I'm going to split the data into smaller chunk that will be used in the experiment.

**Data split strategy:**
- Pick x users
- Pick only merchant and edges connected to remained users

```python
import pandas as pd
```

```python
users_df = pd.read_csv("../data/silver/df_nodes_user_raw.csv")
merchants_df = pd.read_csv("../data/silver/df_nodes_merchant_raw.csv")
edges_pd = pd.read_csv("../data/silver/df_edges_raw.csv")
```

```python
display(users_df)
```

```python
df_user_limited = users_df[users_df['user_id'] < 100]
```

```python
edges_pd = edges_pd.rename(columns={'User': 'user_id'})
```

```python
df_edges_filtered = pd.merge(
    edges_pd,
    df_user_limited[['user_id']], # Bierzemy tylko kolumnę USER_ID z df_USERS
    on='user_id',
    how='inner'
)

df_merchants_filtered = pd.merge(
    merchants_df,
    df_edges_filtered[['merchant_id']].drop_duplicates(), # Bierzemy tylko kolumnę USER_ID z df_USERS
    on='merchant_id',
    how='inner'
)
```

## Save limited datasets

```python
df_edges_filtered.to_csv("../data/silver/u100_df_edges_raw.csv", index=False)
df_user_limited.to_csv("../data/silver/u100_df_nodes_user_raw.csv", index=False)
df_merchants_filtered.to_csv("../data/silver/u100_df_nodes_merchant_raw.csv", index=False)
```
